## Exploratory Data Analysis in Python

Whenever you start on a data project, there are a couple of questions your team needs to ask

1.)  What question are we asking?  What do we want to know, or want to do?

    -It is common for this question to change and evolve during a data project. The initial question will often not be phrased well

2.) What data do we have?   Will it allow us to answer the question?  What is the data like?   What methods of analysis, visualizing, modeling, predicting, etc can we use with this data.




# Exploratory Data Analysis

Typically, at the start of a project, the business team (or scientific, or clinical team, depends on your organization) is going to give you access to the data.   It might be in the form of

    -access to a database run by your organization
    -access to an external database
    -data in a Excel sheet or maybe in a CSV

The first analytic task you will have is to figure out what is in the data

-How big is the dataset- this will influence what tools you can use

-How many data tables does it consist off (for now, we'll stick with just 1)

-For each data table, what are the variables in the table?  How many rows?

    -What types of data are in each column
    -What is the "key" or "identifier variable"-   it may be a composite key of several columns (first name, middle name, last name for  
       example)
    -What is the range of each variable?  What is the min, max, mean, median, standard deviation?
    -What does the histogram of each variable look like?   Is it a bell curve (aka normal or gaussian)
    -For grouping variables or categories (such as customer segments, age ranges, genders, etc) how many distinct categories do you have of   
      each?
    -How much data is missing?   (If any)
    -Which variables are strongly correlated with other variables, indicating a strong positive or negative relationship between them.
    -What do some basic, simple graphs of the data look like?

We need to know this basic background about our data before we really start to do in depth analyses, or produce high quality graphics or complex models.

We may find some bad news during exploratory analysis

    -poor data quality, or missing data
    -data that is inadequate to address the questions we want to ask
    -we may find we need additional data

So we may wind up going back to our team and rethinking how we address the problem, or we may need to look for additional data


## Data Dictionaries

A Data Dictionary is one element of what we call the MetaData associated with a Data Set

A data dictionary defines what all the variables in a data set are, and includes some level of discussion of how the data is measured,
what abbreviations are used,  any coding used, the units in use, maybe even details of how it was collected.

Sadly,  really good data dictionaries are rare.   You will probably have to talk to the "line of business people" or the "domain experts" on
your team about what the data means.

Wikipedia is always a really valuable resource at the start of a project, particularly when you are working in an area you are not familiar with.   It is not an end-all source of knowledge, but it does let you quickly "orient yourself" to some of the terms that come up related to data in a particular field.   Reading a bit quickly on a topic will make it easier for you to talk with business or domain experts about data from their field.    It will give you a chance to look up terms you don't understand before starting the conversation, it allows you to ask better questions.

## What tools should we use?

There is a wide variety of tools we might want to use

1.) Spreadsheets- Excel or Google sheets- these have the advantage of being visual, so we can see all our data at once.   That works for small
data sets,  after about 12 rows and a couple of hundred columns,  it gets overwhelming

2.) Tableau- Tableau has some built-in tools to help with data cleaning and formatting.   I've seen them, but don't use them.  They should 
work fine for smallish data sets.

3.) R- R is easier to use in some ways than Python.   R has a built-in data frame system and really nice graphics.    I use R from relatively 
small data sets, less than 100,000 rows or so.    R has problems with very large data sets.

4.) Python-  Python is the best choice for larger data sets.   We have to use the Pandas library to implement dataframes in Python.    I will start with Python if: 
    a.)  The data set is large.
    b.) I know I will need advanced tools (like deep learning) on a project.   R won't handle really complex, large scale tools as well
         as Python, so if I know I will need Python,  I just start with it.

In addition to Pandas, there are several other dataframe options in Python,  other libraries

    -Dask-   this is a package that allow you to work with dataframes that are too larger to load the entire data set into computer
            memory all at once.    Dask looks like a Pandas dataframe, but uses "blocks" or "chunks" in ways like distributed computing
            does to work with really large data sets.

    -Polars-  Polars is an alternative to Pandas.    Polars has a command structure more like SQL or Tidyverse (from R).    Pandas uses a 
            command structure derived from linear algebra (aka matrix algebra).   Base R, Matlab and Pandas all use a similar "linear 
            algebra" style, while SQL,  Polars and the tidyverse package in R all use methods derived from SQL.

            It is a good idea to learn both approaches (SQL and linear algebra), since they each have strengths and weaknesses.

5.) SQL- You can do data cleaning in SQL.   SQL is not good at producing visualizations, and it doesn't have a lot of analytic tools 
    available, but you can certainly do the basics in SQL.  I find it easy to "pull" data from a database into R or Python and use the wide
    variety of tools in those languages, but if you have really advanced SQL chops, don't hesitate to use it.

## Comments

The function pd.read_csv() has read in our file, and has estimated what type of variable each column is.   The function usually 
does pretty well, but there are ways to over-ride the data type estimates.

"object" in a data frame means text

Rank is a 64 bit integer

Name, Industry, Headquarters, Country, CEO are "objects" or text

Sales, Profit, Assets, Market Value, Founded and Employees are all float objects

"Founded" should actually be a date,   Pandas has a date-time object, we should consider converting this column

"Employees" should really be a float,   I think, but maybe part-time employees count as 1/2 or something?   

```python
# I want a quick description of the data

# The describe function will summarize all the variables for us.

lcdf.describe()
```

```python
# The count listing is the number of value rows for each column.

# It looks like the variables "Founded" and "Employees" might be missing data 

# the isnull() member function will tell us whether or not each entry in a column is a "null" or empty value
# we can combine that with the sum() function that will compute a sum down each column.

missing_counts = lcdf.isnull().sum()

print(missing_counts)
```

This is a pretty cool result for just a couple lines of code

Like I though there are two missing "founded" values and a bunch of misisng "Employee" values,   58 in fact

# Basic graphics

The table is great, but I want some basic graphics as well

```python
# Boxplot

# note that the name "Market Value" has a space in it, this can cause a problem,  try to avoid spaces in names

plt.boxplot(lcdf["Market Value"])
```

Not okay!

This plot is not something we can use, the y-axis is too compressed to be useful.   The handfull of really large companies dominate
the y-axis and most of the interesting stuff is not visible.    

Don't just proceed if you see this problem, you need to fix it

We need to log transform the y-axis so we can see things

```python
# Create the boxplot
plt.boxplot(lcdf["Market Value"])

# Set the y-axis to log scale
plt.yscale('log')

# Display the plot
plt.show()
```

Okay, that is much better, we can now see that the median "Market Value" is around 20 (2 x 10^1)

The max is up around 3000,   I think these must be in Billions of dollars, but don't have the data dictionary

Also notice that I don't have labels on the graphs,  I simply haven't taken the time to do this

For me, these are "working graphics" for my use, or maybe to show quickly to another data scientist

To show to a business audience, you need "presentation graphics" with good labels, and nice formatting and good use of color.

When I'm in a work flow,  I don't care about graphics quality, the default plots give me what I need to know. 

Notice that the problem with the y-axis (it needs to be log transformed) is not just a difference between working and presentation graphics,   we need the transform to effectively read the graph at all.


## Histogram


```python
# Create the boxplot
plt.hist(lcdf["Market Value"])
```

Not okay!

Now we need the log transform on the x-axis, and maybe some more "bins" in the data, or maybe log transforms on both axis

```python
# Create the histogram
plt.hist(lcdf["Market Value"],bins=100,density=True)

# Set the y-axis to log scale
plt.yscale('log')
plt.xscale('log')

# Display the plot
plt.show()
```

I did some adjustment of the x and y axis scales, the number of bins and I set the scaling to plot the density (probability) 
rather than the raw counts of companies in each bin.

I think I would try some other histogram plots to produce a really good plot but this is okay for now

# Biplots

## Pair plot

The pair plot shows a plot of all possible pairwise combinations of variables in the data set along a grid

The plots along the diagonal are the histograms

This is a massive compression of data... helpful only as working graphics during exploratory data analysis

It is a quick way to look at a lot of relationships

see
https://seaborn.pydata.org/generated/seaborn.pairplot.html
for mor options


```python
import seaborn as sns

sns.pairplot(lcdf[["Rank","Sales","Profit",	"Assets","Market Value","Employees"]],diag_kind="kde")
plt.show()
```

## Correlation structure-Heat Map

One of the issues I'm always interested in is the correlation structure in a data set, the tendency for values to change together

We can see this from the biplot of variables relative to one another, or calculate correlation coefficients



```python
# use the built in Pandas function to calculate the correlation of the 6
# variables I am intersted in.  Notice that I am selecting only the 6 float variables out of 12 columns
# in the data frame

corr_matrix = lcdf[["Rank","Sales","Profit","Assets","Market Value","Employees"]].corr(method='pearson')

print(corr_matrix)

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
